Crop the regions of interest and perform OCR on them using Tesseract and Paddle-OCR

```python
import cv2
import os
import datetime
import pytesseract
from paddleocr import PaddleOCR

# Path to the folder containing the images
image_folder = "/content/datasets/Invoice-Processing-1/test/images"

# Create a unique timestamp for the output directory
timestamp = datetime.datetime.now().strftime("%Y%m%d%H%M%S")
output_directory = f"/content/final_folder_{timestamp}/"
os.makedirs(output_directory, exist_ok=True)

# Initialize the YOLO model
model = YOLO('/content/best.pt')

# Initialize Tesseract
pytesseract.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'  # Update with the path to your Tesseract installation

# Initialize PaddleOCR
ocr_paddle = PaddleOCR()

# Iterate over the images in the folder
for image_file in os.listdir(image_folder):
    if image_file.endswith((".jpg", ".png", ".jpeg")):
        image_path = os.path.join(image_folder, image_file)

        # Create a new directory for each image
        image_output_directory = os.path.join(output_directory, f"{image_file}_output/")
        os.makedirs(image_output_directory, exist_ok=True)

        # Load the image
        img = cv2.imread(image_path)

        # Perform object detection
        results = model.predict(img)

        # Iterate over the detection results
        for result in results:
            boxes = result.boxes.cpu().numpy()
            for i, box in enumerate(boxes):
                name = result.names[int(box.cls)]
                r = box.xyxy[0].astype(int)
                print(name)
                print(r)
                crop = img[r[1]:r[3], r[0]:r[2]]
                cv2.imwrite(os.path.join(image_output_directory, str(name) + ".jpg"), crop)

        # Perform OCR with Tesseract on the image
        image_files = [f for f in os.listdir(image_output_directory) if f.endswith((".jpg", ".png", ".jpeg"))]
        ocr_tesseract_results = []

        for image_file in image_files:
            image_path = os.path.join(image_output_directory, image_file)
            image = cv2.imread(image_path)
            ocr_text = pytesseract.image_to_string(image)
            ocr_tesseract_results.append(f"File: {image_file}\nOCR Text (Tesseract):\n{ocr_text}\n")

        # Save the Tesseract OCR results to a text file
        tesseract_output_file = os.path.join(image_output_directory, "ocr_results_tesseract.txt")
        with open(tesseract_output_file, "w") as file:
            file.write("\n\n".join(ocr_tesseract_results))

        # Perform OCR with PaddleOCR on the image
        ocr_paddle_results = []

        for image_file in image_files:
            image_path = os.path.join(image_output_directory, image_file)
            image = cv2.imread(image_path)
            ocr_result = ocr_paddle.ocr(image)
            ocr_text = '\n'.join([word_info[1][0] for line in ocr_result for word_info in line])
            ocr_paddle_results.append(f"File: {image_file}\nOCR Text (PaddleOCR):\n{ocr_text}\n")

        # Save the PaddleOCR results to a text file
        paddle_output_file = os.path.join(image_output_directory, "ocr_results_paddle.txt")
        with open(paddle_output_file, "w") as file:
            file.write("\n\n".join(ocr_paddle_results))

print("OCR process completed!")
```
